```python
pi = pd.read_csv("pi_omp_script.csv", index_col=0,sep=', ', engine='python')
pi_clean = pd.DataFrame({'N':pi.N,
                        'Time':pi.Time})

pi_clean
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for idx, gp in pi_clean.groupby(pi.index):
    gp.plot(x='N',y='Time',ax=ax,label=idx)

#ax = pi_clean.plot(x='N',y='Time',loglog=True, use_index=False, title='Times for Approximations of PI for N(T)')
#ax.legend(pi_clean.index)
ax.set_ylabel('Time')
ax.legend(title='Number of Threads')
ax.set_title('Time for Approximations of Pi of N(T)')
plt.show()
```
