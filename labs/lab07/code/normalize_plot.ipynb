```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
strong = pd.read_csv("normalize_strong.csv",skipinitialspace=True)
```

```python
s_plot = strong.plot(x='n_procs',y=['manual', 'mpi', 'mpi2'])
s_plot.set_title("Strong Scalability for Normalize")
s_plot.set_ylabel("Time(s)")
```

```python
weak = pd.read_csv("normalize_weak.csv",skipinitialspace=True)
w_plot = weak.plot(x='n_procs',y=['manual', 'mpi', 'mpi2'])
w_plot.set_title("Weak Scalability for Normalize")
w_plot.set_ylabel("Time(s)")
```
