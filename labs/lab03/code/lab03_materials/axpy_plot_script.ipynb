```python
# log-log plot for axpy_normal
axpy = pd.read_csv("axpy_normal_script.csv",index_col=0,sep=', ',engine='python')
axpy_random = pd.read_csv("axpy_random_script.csv", index_col=0, sep=', ',engine='python')

axpy.rename(columns={'n':'n', "time":"normal"},inplace=True)
axpy['random'] = axpy_random.time
axpy.normal = 1 * axpy.normal
axpy.random = 1 * axpy.random
axpy.replace(0, np.NaN, inplace=True)
ax = axpy.plot(loglog=True, title='Normal and Random Alogirthm Performance')
ax.set_ylabel("time")
```
